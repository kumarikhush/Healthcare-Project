```python
# data prep - to train the model on training data
#  find the accuracy of prediction of the trained model on the test data

# 70 % of the data as training data and 30 % as the test data

from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
# independent variable
X= df.drop('Outcome', axis=1)
y= df['Outcome']
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train.shape
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# train the model on the training dataset

# inport the logistic regression class


from sklearn.linear_model import LogisticRegression
# create an instance of the Logistic regression class

log_regression=LogisticRegression(max_iter=200)
# log_regression
# fit the logistic regression class on the training data
log_model=log_regression.fit(X_train,y_train)
# use the trained model to predict on the test data
y_prediction=log_model.predict(X_test)
```

```python
y_prediction
```

```python
y_test.values
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_prediction)
```

```python
print(classification_report(y_test,y_prediction))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn_model=knn.fit(X_train,y_train)
y_pred= knn_model.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
# Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree_model=dtree.fit(X_train,y_train)
y_pred= dtree.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
random_forest= RandomForestClassifier()
rand_model=random_forest.fit(X_train,y_train)
y_pred=rand_model.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#Preparing ROC Curve (Receiver Operating Characteristics Curve)
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

# predict probabilities
probs = knn_model.predict_proba(X)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
auc = roc_auc_score(y, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(y, probs)
print("True Positive Rate - {}, False Positive Rate - {} Thresholds - {}".format(tpr,fpr,thresholds))
# plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
```

```python
#Precision Recall Curve for Logistic Regression

from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.metrics import average_precision_score
# predict probabilities
probs = log_model.predict_proba(X)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# predict class values
yhat = log_model.predict(X)
# calculate precision-recall curve
precision, recall, thresholds = precision_recall_curve(y, probs)
# calculate F1 score
f1 = f1_score(y, yhat)
# calculate precision-recall AUC
auc = auc(recall, precision)
# calculate average precision score
ap = average_precision_score(y, probs)
print('f1=%.3f auc=%.3f ap=%.3f' % (f1, auc, ap))
# plot no skill
plt.plot([0, 1], [0.5, 0.5], linestyle='--')
# plot the precision-recall curve for the model
plt.plot(recall, precision, marker='.')
```
